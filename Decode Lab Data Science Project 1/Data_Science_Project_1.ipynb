# **Project 1 - Advanced EDA & Feature Engineering**

**Step 1: Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Step 2: Load Dataset**

```python
df = pd.read_csv('Data Science Project 1.csv')
```

**Step 3: Dataset Information**

```python
print("Shape of Dataset:")
print(df.shape)

print ("\nFirst Five Rows of Dataset:")
print(df.head())

print ("\nLast Five Rows of Dataset:")
print(df.tail())

print("\nColumn of Dataset:")
print(df.columns)

print("\nDataset Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
```

**Step 4: Check Missing Values**

```python
print("Missing Values:")
print(df.isnull().sum())
```

**Step 5: Handle Missing Values**

Since CouponCode is a categorical column, replacing missing values with "No Coupon" is appropriate.

```python
df["CouponCode"].fillna("No Coupon", inplace=True)
print(df.isnull().sum())
```

**Step 6: Convert Date Column**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

**Step 7: Outlier Detection (IQR Method)**

Quantity

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Remove Outliers

```python
df=df[(df['Quantity'] >= lower_bound) & (df['Quantity'] <= upper_bound)]
print(df.shape)
```

Repeat for UnitPrice

```python
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Again Remove Outliers

```python
df = df[(df['UnitPrice'] >= lower_bound) & (df['UnitPrice'] <= upper_bound)]
print(df.shape)
```

Repeat for TotalPrice

```python
Q1 = df['TotalPrice'].quantile(0.25)
Q3 = df['TotalPrice'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
```

Again Remove Outliers

```python
df = df[(df['TotalPrice'] >= lower_bound) & (df['TotalPrice'] <= upper_bound)]
print(df.shape)
```

**Step 8: Feature Engineering**

Feature 1 : Order Month

```python
df['OrderMonth'] = df['Date'].dt.month
```

Feature 2 : Order Year

```python
df['OrderYear'] = df['Date'].dt.year
```

Feature 3 : Average Item Price

```python
df['AverageItemPrice'] = df['TotalPrice'] / df['Quantity']
```

Feature 4 : Coupon Used

```python
df['CouponUsed'] = np.where(df['CouponCode'] == 'No Coupon','No','Yes')
```

Feature 5 : Order Value Category

```python
df['OrderValue'] = pd.cut(df['TotalPrice'],
                          bins=[0,100,500,1000,100000],
                          labels=['Low','Medium','High','Premium'])
```
